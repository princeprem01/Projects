```python
def find_ano2(data):
    t_anomaly=[]
    try:
        col=list(data.select_dtypes(include='number'))
        for i in range(data.shape[1]):
            if data.iloc[:,i].dtype=='int64':

                data_std=statistics.stdev(data.iloc[:,i])
                data_mean=statistics.mean(data.iloc[:,i])
    
                ano_cutoff=5*data_std
    
                lowerl=data_mean-ano_cutoff
                upperl=data_mean+ano_cutoff
    
                anomaly=[]
            
                for outlier in data.iloc[:,i]:
                    if (outlier > upperl) or (outlier < lowerl):
                        anomaly.append(outlier)

                t_anomaly.append((anomaly))
                
            else:
                None
                
#       print(t_anomaly)
    except:
            print("use 1D function, find_ano(data)")
    finally:
            print(dict(zip(col,t_anomaly)))
```

```python
def remove_ano(data):
    t_anomaly=[]
    try:
        col=list(data.select_dtypes(include='number'))
        for i in range(data.shape[1]):
            if (data.iloc[:,i].dtype=='int64') | (data.iloc[:,i].dtype=='float64'):

                data_std=statistics.stdev(data.iloc[:,i])
                data_mean=statistics.mean(data.iloc[:,i])
    
                ano_cutoff=5*data_std #covering 100% of the values
    
                lowerl=data_mean-ano_cutoff
                upperl=data_mean+ano_cutoff
    
                anomaly=[]
            
                for outlier in data.iloc[:,i]:
                    if (outlier > upperl) or (outlier < lowerl):
                        anomaly.append(outlier)

                t_anomaly.append(len(anomaly))
                if (len(anomaly)>0):
                    data=data[~( (data.iloc[:,i]<lowerl) | (data.iloc[:,i]>upperl) )]
                
            else:
                None
                
#       print(t_anomaly)
    except:
            print("use complete dataframe")
            def find_ano(data):
   
                data_std=statistics.stdev(data)
                data_mean=statistics.mean(data)
    
                ano_cutoff=4*data_std #covering 100% of the values
    
                lowerl=data_mean-ano_cutoff
                upperl=data_mean+ano_cutoff
            
                data2=data[~((data<lowerl) | (data>upperl))]
    
                anomaly=[]
    
                for outlier in data:
                    if (outlier > upperl) or (outlier < lowerl):
                        anomaly.append(outlier)
                return data2
            print(find_ano(data))
           
    finally:
#           print(dict(zip(col,t_anomaly)))
            data1=data.copy()
            return data1
```
